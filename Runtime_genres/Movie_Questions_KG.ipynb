```python
# Dependencies
import requests
from config import api_key
import pandas as pd

url = f"http://www.omdbapi.com/?apikey={api_key}&t="
```

```python
#Reading movie title csv file
movie_list = pd.read_csv("highest_grossing_mojo.csv")
movie_list.head()
```

```python
#cleaning df
movie_list_df = movie_list[['Title']]
movie_list_df.count()
```

```python
new_movie_list = movie_list['Title'].to_list()
new_movie_list
```

```python
print(len(new_movie_list))
```

```python
# Loop through list of movies
genre_list = []
runtime_list = []
awards_list = []
director_list = []
movie_name = []
#new_movie_df = pd.DataFrame(columns = ['Title'])
for item in new_movie_list:
    try:
        movie = requests.get(url + item).json()
        movie_name.append(movie["Title"])
        genre_list.append(movie["Genre"])
        runtime_list.append(movie["Runtime"])
        awards_list.append(movie["Awards"])
        director_list.append(movie["Director"])
        #print(f'The genre of {item} was {movie["Genre"]}.')
    except: KeyError
    continue
print(genre_list)        
```

```python
#Making Dictionary for pandas dataframe
data = {
    'movies': movie_name,
    "Genre" : genre_list,
    "Runtime" : runtime_list,
    "Awards" : awards_list,
    "Director" : director_list,
} 
```

```python
movies_df = pd.DataFrame(data)
```
