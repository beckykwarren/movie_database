```python
# Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Store filepath in a variable
csv_file = "Resources/movies_metadata.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
movies_df = pd.read_csv(csv_file)
movies_df.head()
```

```python
# View all column headers
for col in movies_df.columns: 
    print(col) 
```

```python
# Delete unnecessary columns

movies_reduced_df = movies_df.drop(['adult', 'belongs_to_collection', 'genres', 'homepage', 'overview', 'popularity',
                                    'production_companies', 'production_countries','poster_path', 'spoken_languages',
                                    'status', 'tagline', 'video', 'vote_average', 'vote_count'], axis=1)
```

```python
movies_reduced_df.head()
```

```python
# Rearrange columns for efficiency
movies_reduced_df = movies_reduced_df[['id', 'title', 'budget', 'revenue', 'original_title', 'release_date', 'revenue',
                                       'imdb_id', 'original_language']]
```

```python
# View new df 
movies_reduced_df.head()
```

```python
# Export clean df to csv
movies_reduced_df.to_csv("movies_clean.csv")
```
